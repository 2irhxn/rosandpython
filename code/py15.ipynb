# py15: NumPy 字符串操作 (np.char)

`np.char.xxx()` 对字符串数组做**逐元素处理**。

> 新版变化：`numpy.char` 已标为 legacy，新代码优先用 `numpy.strings`；
> NumPy 2.0 移除了 `np.string_` 和 `np.unicode_`，分别用 `np.bytes_` 和 `np.str_`。

```python
import numpy as np
```

## 1. 拼接类函数

```python
# ---- np.char.add：逐元素拼接 ----
a = np.array(["hello", "good"])
b = np.array([" world", " morning"])
print('np.char.add =', np.char.add(a, b))
# 等价于 ["hello" + " world", "good" + " morning"]
```

```python
# ---- np.char.multiply：重复指定次数 ----
a = np.array(["ha", "go"])
print('multiply(a, 3) =', np.char.multiply(a, 3))
# 等价于 ["ha" * 3, "go" * 3] → ['hahaha' 'gogogo']
```

## 2. 大小写处理

```python
a = np.array(["hello", "NumPy", "PYTHON"])

print('upper    =', np.char.upper(a))      # 全部转大写
print('lower    =', np.char.lower(a))      # 全部转小写
print('capitalize =', np.char.capitalize(a))  # 首字母大写，其余小写
print('title    =', np.char.title(a))      # 每个单词首字母大写
```

## 3. 对齐和去除字符

```python
a = np.array(["hi", "numpy"])

# center：居中到指定长度，空位用指定字符填充
print('center(10, "-") =', np.char.center(a, 10, fillchar="-"))
# ['----hi----' '--numpy---']

# ljust / rjust：左对齐 / 右对齐
print('ljust(10, "-") =', np.char.ljust(a, 10, fillchar="-"))
print('rjust(10, "-") =', np.char.rjust(a, 10, fillchar="-"))
```

```python
# ---- strip：去除首尾指定字符 ----
a = np.array(["  hello  ", "---numpy---"])

print('strip 默认空白 =', np.char.strip(a))       # 去首尾空白
print('strip("-")     =', np.char.strip(a, "-"))  # 去首尾-
print('lstrip("-")    =', np.char.lstrip(a, "-")) # 只去左边
print('rstrip("-")    =', np.char.rstrip(a, "-")) # 只去右边
```

## 4. 分割和换行处理

```python
# ---- split：按分隔符分割 ----
a = np.array(["a,b,c", "1,2,3"])
print('split(",") =')
print(np.char.split(a, sep=","))
# [list(['a', 'b', 'c']) list(['1', '2', '3'])]

# splitlines：按换行符分割
b = np.array(["hello\nworld", "python\nnumpy"])
print('splitlines =')
print(np.char.splitlines(b))
```

## 5. 替换和连接

```python
# ---- replace：替换子串 ----
a = np.array(["hello world", "hello numpy"])
print('replace =', np.char.replace(a, "hello", "hi"))
# ['hi world' 'hi numpy']
```

```python
# ---- join：对每个字符串内部的字符进行连接 ----
a = np.array(["abc", "123"])
print('join("-", a) =', np.char.join("-", a))
# ['a-b-c' '1-2-3']

# 注意：这不是把数组元素合成一个字符串
# 合成一个字符串用 Python 的 " ".join(arr)
```

## 6. 编码和解码

```python
# ---- encode / decode ----
a = np.array(["中文", "hello"])

b = np.char.encode(a, "utf-8")
print('encode =', b)
print('dtype =', b.dtype)  # |S6

c = np.char.decode(b, "utf-8")
print('decode =', c)
```

## 7. 字符串数组的 dtype

```python
# Unicode 字符串： dtype 是 <U + 最大长度
a = np.array(["hi", "hello", "numpy"])
print('a.dtype =', a.dtype)  # <U5（最长字符串5个字符）

# bytes 字节字符串： dtype 是 |S + 最大长度
b = np.array([b"hi", b"hello"])
print('b.dtype =', b.dtype)  # |S5

# astype 转换字符串类型
c = a.astype("U10")
print('astype U10 =', c.dtype)
```

## 8. 实际应用示例

```python
# 批量处理文件名
names = np.array([" img_001.jpg ", " img_002.jpg ", " img_003.jpg "])

names = np.char.strip(names)                    # 去空格
names = np.char.replace(names, ".jpg", ".png")  # 改后缀
print('处理后 =', names)
# ['img_001.png' 'img_002.png' 'img_003.png']
```

```python
# 批量格式化标签
labels = np.array(["cat", "dog", "bird"])
counts = np.array([5, 3, 8])

# 用 add 拼接
result = np.char.add(labels, np.char.multiply(": ", 1))
result = np.char.add(result, counts.astype(str))
print('格式化标签 =', result)
# ['cat: 5' 'dog: 3' 'bird: 8']
```

## 9. 要点总结

1. `np.char.xxx()` 是对字符串数组做**逐元素处理**
2. 字符串数组 dtype：`<U` (Unicode) 或 `|S` (bytes)
3. 优先用 `np.char` 做批量字符串操作，比循环更高效
4. 普通文本处理用 Python 字符串方法 / pandas `.str` 更常见
