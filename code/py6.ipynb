## dict 字典

dict 用键查找值，适合表达“名字 -> 分数”这类映射关系。键必须是不可变且可哈希的对象。


```python
#dict
names = ['Michael', 'Bob', 'Tracy']
scores = [95, 75, 85]

d = {'Michael': 95, 'Bob': 75, 'Tracy': 85}
print(d['Michael'])
d['Adam'] = 67
print(d)
d['Jack'] = 90
d['Jack'] = 88 #覆盖Jack原来的值
print(d)
d.pop('Bob') #删除Bob的键值对
print(d)
```

## set 集合

set 自动去重，适合成员判断、交集、并集等集合运算。


```python
#set
s = set([1, 2, 3])
print(s)
s.add(4)
print(s)
s.remove(2)
print(s)
s.discard(5) #不报错，如果不存在则不会报错
print(s)

s = set([1, 2, 3])
print(s)
s = {1, 1, 2, 2, 4, 3, 3,}
print(s)

p = {1,2,6,'apple', 'banana', 'cherry'}
print(s & p) #交集
print(s | p) #并集
```
