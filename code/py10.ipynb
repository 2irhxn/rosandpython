## 模块元信息

模块顶部可以放文档字符串、作者信息和 import；被导入时这些定义会被加载。


```python
' a test module '

__author__ = 'Sen'

import sys
```

## 读取命令行参数

sys.argv 保存命令行参数；在 notebook 中它通常来自 Jupyter 内核环境，和命令行运行脚本不同。


```python
def test():
    args = sys.argv
    if len(args)==1:
        print('Hello, world!')
    elif len(args)==2:
        print('Hello, %s!' % args[1])
    else:
        print('Too many arguments!')
```

## 内部函数与对外函数

以下划线开头的函数表示“内部使用”的约定，greeting() 是对外调用的函数。


```python
def _private_1(name):
    return 'Hello, %s' % name

def _private_2(name):
    return 'Hi, %s' % name

def greeting(name):
    if len(name) > 3:
        return _private_1(name)
    else:
        return _private_2(name)
```

## 主程序入口

只有直接运行脚本时 __name__ 才等于 "__main__"；被 import 时这里的代码不会自动执行。


```python
if __name__=='__main__':
    test()
    print(greeting('Sen'))
```
