# py19: NumPy 矩阵库与线性代数

本 notebook 演示矩阵库 (numpy.matrix) 和线性代数 (numpy.linalg) 常用函数。

```python
import numpy as np
```

## 1. 矩阵库 (numpy.matrix)

> 注意：`numpy.matrix` 是 ndarray 的子类，专为矩阵运算设计。
> 新代码建议优先使用 `np.array` + `@` 运算符，但了解 matrix 有助于阅读旧代码。

```python
# ---- 创建矩阵 ----
a = np.matrix('1 2 3; 4 5 6; 7 8 9')
print('矩阵 a =')
print(a)
print('类型:', type(a))

# 从列表创建
b = np.matrix([[1, 2], [3, 4]])
print('矩阵 b =')
print(b)
```

```python
# ---- 矩阵运算（* 直接做矩阵乘法） ----
a = np.matrix([[1, 2], [3, 4]])
b = np.matrix([[5, 6], [7, 8]])

print('a * b (矩阵乘法) =')
print(a * b)  # matrix 的 * 是矩阵乘法
# [[19 22]
#  [43 50]]

# ndarray 的 * 是逐元素乘法，需要用 @ 或 dot()
a_arr = np.array([[1, 2], [3, 4]])
b_arr = np.array([[5, 6], [7, 8]])
print('ndarray @ =')
print(a_arr @ b_arr)
```

```python
# ---- 矩阵特有属性 ----
a = np.matrix([[1, 2], [3, 4]])

print('a * a =')              # 矩阵幂
print(a * a)

print('a.I =')               # 逆矩阵
print(a.I)

print('a.T =')               # 转置
print(a.T)

print('a.H =')               # 共轭转置（实数矩阵同 T）
print(a.H)
```

## 2. 线性代数 (numpy.linalg)

常用函数一览：

| 函数 | 说明 |
|---|---|
| `linalg.det()` | 行列式 |
| `linalg.inv()` | 逆矩阵 |
| `linalg.solve()` | 解线性方程组 |
| `linalg.eig()` | 特征值和特征向量 |
| `linalg.svd()` | 奇异值分解 |
| `linalg.norm()` | 向量/矩阵范数 |
| `linalg.lstsq()` | 最小二乘解 |

```python
# ---- det：行列式 ----
a = np.array([[1, 2], [3, 4]])
print('a =')
print(a)
print('det(a) =', np.linalg.det(a))  # 1*4-2*3 = -2
```

```python
# ---- inv：逆矩阵 ----
a = np.array([[1, 2], [3, 4]])
a_inv = np.linalg.inv(a)
print('inv(a) =')
print(a_inv)

# 验证：a @ a_inv ≈ 单位矩阵
print('a @ inv(a) ≈ I =')
print(np.round(a @ a_inv, 10))  # 四舍五入消除浮点误差
```

```python
# ---- solve：解线性方程组 Ax = b ----
# 2x + 3y = 8
# 3x + 4y = 11
A = np.array([[2, 3],
              [3, 4]])
b = np.array([8, 11])

x = np.linalg.solve(A, b)
print('解 x =', x)  # [1. 2.]
print('验证 A @ x =', A @ x)  # [8. 11.]
```

```python
# ---- eig：特征值和特征向量 ----
a = np.array([[4, -2],
              [1,  1]])
print('a =')
print(a)

vals, vecs = np.linalg.eig(a)
print('特征值 =', vals)
print('特征向量 =')
print(vecs)

# 验证：a @ v = λ * v
print('验证 a @ vecs[:,0] =', a @ vecs[:,0])
print('         vals[0] * vecs[:,0] =', vals[0] * vecs[:,0])
```

```python
# ---- svd：奇异值分解 ----
a = np.array([[1, 2, 3],
              [4, 5, 6]])
print('a = shape', a.shape)
print(a)

U, S, Vt = np.linalg.svd(a)
print('\nU (左奇异向量) shape =', U.shape)
print(U)
print('\nS (奇异值) =', S)
print('\nVt (右奇异向量) shape =', Vt.shape)
print(Vt)

# 验证：U @ diag(S) @ Vt ≈ a
print('\n重建 ≈ a:')
print(np.round(U @ np.diag(S) @ Vt, 10))
```

```python
# ---- norm：范数 ----
a = np.array([3, 4])
print('向量 a =', a)
print('L2 范数 (欧氏距离) =', np.linalg.norm(a))  # 5.0
print('L1 范数 =', np.linalg.norm(a, ord=1))       # 7
print('无穷范数 =', np.linalg.norm(a, ord=np.inf))  # 4

# 矩阵范数
m = np.array([[1, 2], [3, 4]])
print('矩阵 Frobenius 范数 =', np.linalg.norm(m))
print('矩阵 L1 范数 =', np.linalg.norm(m, ord=1))  # 列和最大值
```

```python
# ---- lstsq：最小二乘解 ----
# 拟合 y = kx + b
x = np.array([1, 2, 3, 4, 5])
y = np.array([2.1, 4.0, 5.9, 8.1, 10.0])

# 构造增广矩阵 [x, 1]
A = np.vstack([x, np.ones(len(x))]).T
print('A =')
print(A)

# 最小二乘解
k, b = np.linalg.lstsq(A, y, rcond=None)[0]
print(f'拟合结果: y = {k:.2f}x + {b:.2f}')
print('预测值 =', k * x + b)
```

## 3. 总结

| 函数 | 说明 | 示例 |
|---|---|---|
| `np.linalg.det(A)` | 行列式 | `det([[1,2],[3,4]])` → -2 |
| `np.linalg.inv(A)` | 逆矩阵 | `A @ inv(A)` ≈ I |
| `np.linalg.solve(A,b)` | 解 Ax=b | 直接求解 |
| `np.linalg.eig(A)` | 特征值分解 | vals, vecs = eig(A) |
| `np.linalg.svd(A)` | SVD 分解 | U, S, Vt = svd(A) |
| `np.linalg.norm(A)` | 范数 | 欧氏距离、矩阵范数 |
| `np.linalg.lstsq(A,b)` | 最小二乘 | 回归拟合 |

**优先使用 `np.array` + `@` 运算符**，而非 `np.matrix`。
