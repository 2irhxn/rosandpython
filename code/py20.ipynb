# py20: NumPy IO 与 Matplotlib 基础

本 notebook 演示 NumPy 文件读写 (save/load) 和 Matplotlib 数据可视化。

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. NumPy 文件 IO

```python
# ---- np.save / np.load：.npy 格式（推荐） ----
a = np.array([[1, 2, 3],
              [4, 5, 6]])

# 保存为 .npy 文件（自动压缩，保留 dtype）
np.save('data.npy', a)
print('已保存 data.npy')

# 读取
b = np.load('data.npy')
print('读取结果 =')
print(b)
```

```python
# ---- np.savez：保存多个数组到 .npz 文件 ----
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

np.savez('multi.npz', arr1=x, arr2=y)
print('已保存 multi.npz')

# 读取（按名称访问）
data = np.load('multi.npz')
print('arr1 =', data['arr1'])
print('arr2 =', data['arr2'])
print('所有文件名:', list(data.files))
```

```python
# ---- np.savetxt / np.loadtxt：文本格式 ----
a = np.array([[1.5, 2.3, 3.1],
              [4.7, 5.6, 6.8]])

# 保存为 CSV（逗号分隔，保留2位小数）
np.savetxt('data.csv', a, delimiter=',', fmt='%.2f')
print('已保存 data.csv')

# 读取 CSV
b = np.loadtxt('data.csv', delimiter=',')
print('读取结果 =')
print(b)
```

```python
# ---- np.genfromtxt：处理复杂文本 ----
# 模拟带缺失值的 CSV
import io
csv_data = """1,2,3
4,,6
7,8,9"""

data = np.genfromtxt(io.StringIO(csv_data), delimiter=',', filling_values=0)
print('genfromtxt 处理缺失值 =')
print(data)  # 缺失值自动填 0
```

## 2. Matplotlib 基础

Matplotlib 是 Python 最常用的绑图库，与 NumPy 无缝配合。

```python
# ---- 折线图 ----
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('三角函数')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ---- 散点图 ----
np.random.seed(42)
x = np.random.randn(50)
y = 2 * x + np.random.randn(50) * 0.5
colors = np.random.rand(50)
sizes = np.abs(np.random.randn(50)) * 100

plt.figure(figsize=(6, 5))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.colorbar(label='颜色值')
plt.title('散点图')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# ---- 柱状图 ----
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 33]

plt.figure(figsize=(6, 4))
plt.bar(categories, values, color=['red', 'green', 'blue', 'orange', 'purple'])
plt.title('柱状图')
plt.ylabel('值')
plt.show()
```

```python
# ---- 直方图 ----
data = np.random.normal(100, 15, 1000)  # 均值100，标准差15

plt.figure(figsize=(6, 4))
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(data.mean(), color='red', linestyle='--', label=f'均值={data.mean():.1f}')
plt.title('直方图')
plt.xlabel('值')
plt.ylabel('频数')
plt.legend()
plt.show()
```

```python
# ---- 子图 subplot ----
x = np.linspace(0, 2 * np.pi, 100)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# 左上：sin
axes[0, 0].plot(x, np.sin(x), 'r-')
axes[0, 0].set_title('sin(x)')

# 右上：cos
axes[0, 1].plot(x, np.cos(x), 'b-')
axes[0, 1].set_title('cos(x)')

# 左下：tan
axes[1, 0].plot(x, np.tan(x), 'g-')
axes[1, 0].set_ylim(-5, 5)
axes[1, 0].set_title('tan(x)')

# 右下：指数
axes[1, 1].plot(x, np.exp(x/3), 'm-')
axes[1, 1].set_title('exp(x/3)')

plt.tight_layout()
plt.show()
```

```python
# ---- 热力图 imshow ----
data = np.random.rand(10, 10)

plt.figure(figsize=(6, 5))
plt.imshow(data, cmap='hot', interpolation='nearest')
plt.colorbar(label='值')
plt.title('热力图')
plt.show()
```

```python
# ---- 保存图片 ----
x = np.linspace(0, 2 * np.pi, 100)
plt.figure(figsize=(6, 4))
plt.plot(x, np.sin(x))
plt.title('保存示例')
plt.savefig('plot.png', dpi=150, bbox_inches='tight')
print('图片已保存 plot.png')
plt.show()
```

## 3. NumPy IO 总结

| 函数 | 格式 | 说明 |
|---|---|---|
| `np.save(f, a)` | .npy | 保存单个数组（二进制） |
| `np.load(f)` | .npy | 读取数组 |
| `np.savez(f, **a)` | .npz | 保存多个数组 |
| `np.savetxt(f, a)` | .csv/txt | 保存为文本 |
| `np.loadtxt(f)` | .csv/txt | 读取文本 |
| `np.genfromtxt(f)` | .csv/txt | 处理缺失值的文本读取 |

**推荐使用 `.npy` / `.npz` 格式**，速度快且保留 dtype 信息。
